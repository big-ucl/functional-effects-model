```python
def add_NA(df, col):
    df.loc[df[col] < 0, col] = -1
    return df
```

```python
na_one_five_mapping = {
    -1: "NA",
    1: "yes",
    5: "no",
}
```

```python
def preprocess_one_five_NA(df, cols):
    for col in cols:
        df = add_NA(df, col)
        df[col] = df[col].replace(na_one_five_mapping)

    return df
```

```python
def preprocess_cont_NA(df, cols):
    for col in cols:
        df = add_NA(df, col)
        df[col + "_NA"] = (df[col] == -1).astype(int)
        df[col] = df[col].replace(-1, 0)
    
    return df
```

```python
def preprocess_cat_NA(df, cols, mapping):
    for col in cols:
        df = add_NA(df, col)
        df[col] = df[col].replace(mapping[col])

    return df
```
