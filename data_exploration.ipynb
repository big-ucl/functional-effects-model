```python
def remove_non_num_chars(string, to_dtype=None):
    if to_dtype == int:
        return to_dtype(''.join([c for c in string if c.isdigit() or c == "-"]))
    elif to_dtype == float:
        return to_dtype(''.join([c for c in string if c.isdigit() or c == "-" or c == "."]))
    else:
        return ''.join([c for c in string if c.isdigit()])
```

```python
df["int_version"].apply(remove_non_num_chars, to_dtype=int)
```
