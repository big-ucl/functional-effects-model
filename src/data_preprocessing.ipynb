```python
def add_NA(df, col):
    df.loc[df[col] < 0, col] = -1
    return df
```

```python
na_one_five_mapping = {
    -1: "NA",
    1: "yes",
    5: "no",
}
```

```python
def preprocess_one_five_NA(df, cols):
    for col in cols:
        df = add_NA(df, col)
        df[col] = df[col].replace(na_one_five_mapping)

    return df
```

```python
def preprocess_cont_NA(df, cols):
    for col in cols:
        df = add_NA(df, col)
        df[col] = df[col].replace(-1, 0)
    
    return df
```

```python
def preprocess_cat_NA(df, cols, mapping):
    for col in cols:
        df = add_NA(df, col)
        df[col] = df[col].replace(mapping[col])

    return df
```

```python
def remove_columns(df, cols, threshold=0.1):
    col_to_remove = []
    for col in cols:
        if (df[col] < 0).sum() / df.shape[0] > threshold:
            col_to_remove.append(col)
    df = df.drop(columns=col_to_remove)
    print(f"Removed {len(col_to_remove)} columns with more than {threshold * 100}% missing values")
    return df, col_to_remove
```

```python
def remove_rows(df, cols):
    num_rows_removed = 0
    for col in cols:
        num_rows_removed += df[df[col] < 0].shape[0]
        df = df[df[col] >= 0]
    print(f"Removed {num_rows_removed} rows with missing values")
    return df
```

```python
features = [c for c in df_preprocessed.columns if c not in ["mergeid", "hhid", "coupleid", "country"]]

#missing values
df_preprocessed, cols_removed = remove_columns(df_preprocessed, features, threshold=0.1)

new_features = [c for c in df_preprocessed.columns if c not in ["mergeid", "hhid", "coupleid", "country"]]
df_preprocessed = remove_rows(df_preprocessed, new_features)
```

# Modelling choice summary

- Columns deemed irrelevant:
    * wavepart: irrelevant, 
    * "int_year": irrelevant 
    * "int_month": irrelevant, 
    * "country": used with dummies instead and ISO code,
    * "language": assumed to be mostly contained within country, too many Nans,
    * "dn002_mod"; month of birth, assumed to be contained within age
    * "dn003_mod": year of birth, contained within age, 
    * "birth_country": contained with citizenship 
    * "iv009_mod": location of the interview building, irrelevant,
    * "q34_re": religious denomination, only wave 1
    * "eduyears_mod": years of education, not asked in wave 1 and 3 so removed
    * "int_partner": interview of partner 
    * "sp003-9_X": who gave help, irrelevant,
    * casp": quality of life, lots of NA
    * "euroX": contained in the dependant variable euroD
    * "bfi10_X": too many NA, not relevant as well,
    * "ep0X_mod": only for employed persons, too many NA,
    * "adlwa": use of adla instead,
    * "iadla": use of iadlza instead,
    * "bmi2": use of raw bmi instead,
    * "numeracy_1", "numeracy_2": different questions not asked to the same person
    * "income_pct_wX": use of raw income instead
- Removing columns with more than 10% of missing values
- Indicator variable for missing value if the columns applies to only some persons
- Removing remaining missing values observations
- Normalising one categories for dummies
- Drop wave 3 and 7 as it was a different survey
- Drop variable on respondent history as it was only asked in some waves
- Removing NA for the dependent variable
